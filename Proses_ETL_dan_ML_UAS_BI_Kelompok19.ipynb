```python
df = pd.read_csv('/content/drive/MyDrive/UAS BI/SuperStoreOrders.csv')

df

```

**TRANSFORM - DATA CLEANING**

1. Hapus data kosong

```python
df.dropna(inplace=True)
```

2. Hapus data duplikat

```python
df.drop_duplicates(inplace=True)
```

Konversi Tipe Data

```python
# KONVERSI TANGGAL
df['order_date'] = pd.to_datetime(
    df['order_date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)

df['ship_date'] = pd.to_datetime(
    df['ship_date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)

# KONVERSI NUMERIK
numeric_columns = [
    'sales',
    'quantity',
    'discount',
    'profit',
    'shipping_cost',
    'year'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )

df.dropna(inplace=True)
```

**DIMENSION TABLE**

1. Dim Customer

```python
dim_customer = df[[
    'customer_name',
    'segment'
]].drop_duplicates().reset_index(drop=True)

dim_customer['customer_id'] = range(1, len(dim_customer)+1)
```

2. Dim Product

```python
dim_product = df[[
    'product_id',
    'category',
    'sub_category',
    'product_name'
]].drop_duplicates().reset_index(drop=True)
```

3. Dim Location

```python
dim_location = df[[
    'state',
    'country',
    'market',
    'region'
]].drop_duplicates().reset_index(drop=True)

dim_location['location_id'] = range(1, len(dim_location)+1)
```

4. Dim Time

```python
dim_time = df[[
    'order_date',
    'ship_date',
    'year'
]].drop_duplicates().reset_index(drop=True)

dim_time['time_id'] = range(1, len(dim_time)+1)
```

Merge ID ke Dataframe Utama

```python
# customer_id
df = df.merge(
    dim_customer,
    on=['customer_name', 'segment'],
    how='left'
)

# location_id
df = df.merge(
    dim_location,
    on=['state', 'country', 'market', 'region'],
    how='left'
)

# time_id
df = df.merge(
    dim_time,
    on=['order_date', 'ship_date', 'year'],
    how='left'
)
```

# **MACHINE LEARNING**

```python
from sklearn.cluster import KMeans
```

**MODEL**

```python
# ==================== 1. FITUR CLUSTERING & MODEL K-MEANS ====================
X = df[['sales', 'profit']]

model = KMeans(
    n_clusters=3,
    random_state=42
)

# Training + masukkan hasil cluster langsung ke dataframe utama (df)
df['cluster'] = model.fit_predict(X)

cluster_mapping = {
    0: 'Rendah',
    1: 'Unggulan',
    2: 'Menengah'
}

df['cluster_label'] = df['cluster'].map(cluster_mapping)
```

```python
fact_sales = df[[
    'order_id',
    'customer_id',
    'product_id',
    'location_id',
    'time_id',
    'sales',
    'quantity',
    'discount',
    'profit',
    'shipping_cost',
    'cluster'
]]
```
